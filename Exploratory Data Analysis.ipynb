<h4>Under standing the dataset to explore how the data is present in the dataset  and if there is a need of creating some aggregated tables that can help with: </h4>

<li>Product Pricing Optimization</li>

```python
import pandas as pd
import sqlite3
```

```python
#creating database connection
conn=sqlite3.connect('inventory.db')
```

```python
#checking tables present in the database
tables=pd.read_sql_query("SELECT name From sqlite_master WHERE type='table'",conn)
tables
```

```python
for table in tables['name']:
    print('-'*50, f'{table}','-'*50)
    print('Count of records:',pd.read_sql(f"select count(*) as count from {table}",conn)['count'].values[0])
    display(pd.read_sql(f"select * from {table} limit 5",conn))
```

```python
pd.read_sql('SELECT COUNT(*) FROM `sales`', conn)
```

```python
for table in tables['name']:
    print('-'*50, table, '-'*50)
    
    # Count rows
    count_df = pd.read_sql(f'SELECT COUNT(*) AS count FROM "{table}"', conn)
    print('Count of records:', count_df['count'].iloc[0])
    
    # Fast preview using rowid
    try:
        preview = pd.read_sql(
            f'SELECT * FROM "{table}" ORDER BY rowid LIMIT 5',
            conn
        )
        display(preview)
    except Exception as e:
        print("Could not preview table:", e)
```

```python
pd.read_sql('SELECT COUNT(*) AS total FROM "sales"', conn)
```
